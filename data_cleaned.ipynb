```python
import os
import pandas as pd

# Define the paths
original_path = '/root/autodl-tmp/Stockformer/Stockformer_run/us_alpha158_21-23'
processed_path = '/root/autodl-tmp/Stockformer/Stockformer_run/us_alpha158_21-23_processed'

# Create the processed directory if it doesn't exist
if not os.path.exists(processed_path):
    os.makedirs(processed_path)

# Load LABEL0.csv file
label0_path = os.path.join(original_path, 'LABEL0.csv')
label0_df = pd.read_csv(label0_path, index_col=0)

# Find all columns that are entirely NaN
nan_columns = label0_df.columns[label0_df.isna().all()].tolist()

# Remove these columns from LABEL0.csv
label0_df.drop(columns=nan_columns, inplace=True)

label0_df = label0_df.fillna(0)

label0_df = label0_df[1:-2]

# Save the cleaned LABEL0.csv file in the processed directory
label0_df.to_csv(os.path.join(processed_path, 'LABEL0.csv'))

# Iterate over all other CSV files in the original directory
files = os.listdir(original_path)
for file in files:
    if file.endswith('.csv') and file != 'LABEL0.csv':
        file_path = os.path.join(original_path, file)
        
        # Load the data from CSV file
        df = pd.read_csv(file_path, index_col=0)
        
        # Remove the NaN columns
        df.drop(columns=nan_columns, inplace=True)

        df = df.fillna(0)

        df = df[1:-2]
        
        # Save the cleaned file in the processed directory
        df.to_csv(os.path.join(processed_path, file))
```

```python
import os
import pandas as pd

# Define the path for the processed files
processed_path = '/root/autodl-tmp/Stockformer/Stockformer_run/us_alpha158_21-23_processed'

# Initialize a dictionary to hold the dimensions of each file
file_dimensions = {}

# Iterate over all CSV files in the processed directory
files = os.listdir(processed_path)
for file in files:
    if file.endswith('.csv'):
        file_path = os.path.join(processed_path, file)
        
        # Load the CSV file
        df = pd.read_csv(file_path)
        
        # Store the dimensions of the dataframe
        file_dimensions[file] = df.shape

# Check if all dimensions are the same
dimensions_check = len(set(file_dimensions.values())) == 1

# Print the result
if dimensions_check:
    print("All CSV files have the same dimensions.")
else:
    print("CSV files have different dimensions.")

# Optionally, print the dimensions of each file
for file, dims in file_dimensions.items():
    print(f"{file}: {dims}")
```
